Uma marcenaria fabrica mesas e cadeiras. Cada mesa vendida gera um lucro de R\$ 70, e cada cadeira, R\$ 30. A produção desses móveis é limitada por três recursos: tempo de trabalho, quantidade de madeira disponível e capacidade de armazenamento.

As limitações são as seguintes:

- Cada mesa exige 5 horas de trabalho, e cada cadeira, 2 horas. A marcenaria dispõe de 60 horas semanais.

- Cada mesa consome 8 metros de madeira, e cada cadeira, 4 metros. O estoque semanal de madeira é de 96 metros.

- Por questões de espaço no depósito, a produção total de mesas e cadeiras não pode ultrapassar 16 unidades por semana.

Quantas mesas e cadeiras devem ser produzidas semanalmente para maximizar o lucro, considerando que os móveis só podem ser produzidos em quantidades inteiras?

```python
#!pip install ortools
```

```python
# entrada de dados
n = 2
lucro = [70, 30]
coeficientes = [[5, 8], [2, 4], [1,1]]
limitante = [60, 96, 16]
```

```python
from ortools.linear_solver import pywraplp #importa a biblioteca
```

```python
solver = pywraplp.Solver.CreateSolver('SCIP') #declara o solver: SCIP para PLI e GLOP para PL
infinity = solver.infinity() #declara um valor muito grande (infinity) caso seja necessário usar
```

```python
x = []
x.append(solver.IntVar(0.0, infinity, 'mesa'))
x.append(solver.IntVar(0.0, infinity, 'cadeira'))
```

```python
objetivo = solver.Objective()
for i in range(n):
  objetivo.SetCoefficient(x[i], lucro[i])
objetivo.SetMaximization()
```

```python
for i in range(len(coeficientes)):
  restricao = solver.Constraint(-infinity, limitante[i])
  for j in range(len(coeficientes[i])):
    restricao.SetCoefficient(x[j], coeficientes[i][j])
```

```python
status = solver.Solve()
if status == pywraplp.Solver.OPTIMAL:
  print('Solução:')
  print('Valor da função objetivo =', solver.Objective().Value())
  for i in range(n):
    print(x[i].name(), '=', x[i].solution_value())
```
